```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('hbr_caso_cliente_responde_oferta.csv')

print(df.head(5)) 
```

```python
df.info()
```

```python
#Variables categoricas
df_var_cat = ['cliente_id','genero','nivel_educacion','region']
```

```python
# Variables numericas
df_temp = df.drop(columns=df_var_cat)
df_num = df_temp.select_dtypes(include=['number'])
```

```python
cantidad_clientes = df['cliente_id'].nunique()
print(f"Cantidad de clientes únicos: {cantidad_clientes}")
```

#### Compisición de clientes por género

```python
plt.figure(figsize=(6,4))
sns.countplot(data=df, x='genero', palette='pastel')
plt.title('Cantidad de clientes por género')
plt.xlabel('Género')
plt.ylabel('Cantidad')
plt.show()
```

#### Compisición de clientes por educación

```python
plt.figure(figsize=(8,4))
sns.countplot(data=df, x='nivel_educacion', palette='muted')
plt.title('Cantidad de clientes por nivel de educación')
plt.xlabel('Nivel de educación')
plt.ylabel('Cantidad')
plt.xticks(rotation=45)
plt.show()
```

#### Compisición de clientes por región

```python
plt.figure(figsize=(10,4))
sns.countplot(data=df, x='region', palette='deep')
plt.title('Cantidad de clientes por región')
plt.xlabel('Región')
plt.ylabel('Cantidad')
plt.xticks(rotation=45)
plt.show()
```

```python

app_counts = df['usa_app'].value_counts()

plt.figure(figsize=(5,5))
plt.pie(app_counts, labels=app_counts.index, autopct='%1.1f%%', startangle=90, colors=['#66b3ff','#ff9999'])
plt.title('Uso de la App')
plt.axis('equal')  # Mantiene la forma circular
plt.show()
```

```python
web_counts = df['usa_web'].value_counts()

plt.figure(figsize=(5,5))
plt.pie(web_counts, labels=web_counts.index, autopct='%1.1f%%', startangle=90, colors=['#99ff99','#ffcc99'])
plt.title('Uso del Sitio Web')
plt.axis('equal')
plt.show()
```

```python
respondio_oferta_counts = df['respondio_oferta'].value_counts()

plt.figure(figsize=(5,5))
plt.pie(respondio_oferta_counts, labels=respondio_oferta_counts.index, autopct='%1.1f%%', startangle=90, colors=["#ac99ff","#cfa13c"])
plt.title('Tasa Respuesta Oferta')
plt.axis('equal')
plt.show()
```
